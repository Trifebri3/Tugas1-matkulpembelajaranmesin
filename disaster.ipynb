```python
# IMPORT LIBRARY YANG DIPERLUKAN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Konfigurasi tampilan visualisasi
plt.style.use('classic')
%matplotlib inline

# ============================================
# 1. MEMBACA DATASET
# ============================================

print("="*50)
print("MEMBACA DATASET DISASTER.CSV")
print("="*50)

# Membaca file CSV
df = pd.read_csv('disaster.csv')

# Menampilkan 5 data teratas
print("\n5 Data Pertama:")
print(df.head())

# Menampilkan informasi dataset
print("\nInformasi Dataset:")
print(df.info())

# Menampilkan statistik deskriptif
print("\nStatistik Deskriptif:")
print(df.describe())

# ============================================
# 2. DATA CLEANING & PREPROCESSING
# ============================================

print("\n" + "="*50)
print("DATA CLEANING & PREPROCESSING")
print("="*50)

# Cek missing values
print("\nCek Missing Values:")
print(df.isnull().sum())

# Cek duplicate data
print("\nCek Data Duplikat:")
print(f"Jumlah duplikat: {df.duplicated().sum()}")

# Statistik lengkap termasuk kolom kategorikal
print("\nStatistik Lengkap (Include Object):")
print(df.describe(include='all'))

# ============================================
# 3. EKSPLORASI DATA (EDA)
# ============================================

print("\n" + "="*50)
print("EKSPLORASI DATA (EDA)")
print("="*50)

# Distribusi Umur
print("\nDistribusi Umur Responden:")
print(df['age'].value_counts().sort_index())

# Distribusi Gender
print("\nDistribusi Gender:")
print(df['gender'].value_counts())
print(df['gender'].value_counts(normalize=True) * 100)

# Distribusi Pendidikan
print("\nDistribusi Tingkat Pendidikan:")
print(df['education_level'].value_counts())

# Distribusi Jenis Bencana
print("\nDistribusi Jenis Bencana:")
print(df['disaster_type'].value_counts())

# ============================================
# 4. VISUALISASI DATA
# ============================================

print("\n" + "="*50)
print("VISUALISASI DATA")
print("="*50)
print("Menampilkan grafik...")

# Membuat figure dengan subplots
fig = plt.figure(figsize=(20, 16))

# 4.1 Distribusi Umur
ax1 = plt.subplot(3, 3, 1)
df['age'].hist(bins=20, edgecolor='black', alpha=0.7, color='blue')
plt.title('Distribusi Umur Responden', fontsize=12, fontweight='bold')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')
plt.grid(True, alpha=0.3)

# 4.2 Distribusi Gender
ax2 = plt.subplot(3, 3, 2)
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=['lightblue', 'pink'])
plt.title('Distribusi Gender', fontsize=12, fontweight='bold')

# 4.3 Distribusi Pendidikan
ax3 = plt.subplot(3, 3, 3)
df['education_level'].value_counts().plot(kind='bar', color='green', edgecolor='black')
plt.title('Distribusi Tingkat Pendidikan', fontsize=12, fontweight='bold')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)

# 4.4 Distribusi Jenis Bencana
ax4 = plt.subplot(3, 3, 4)
df['disaster_type'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title('Distribusi Jenis Bencana', fontsize=12, fontweight='bold')
plt.xlabel('Jenis Bencana')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)

# 4.5 Training Received vs Not
ax5 = plt.subplot(3, 3, 5)
training_counts = df['training_received'].value_counts()
plt.pie(training_counts.values, labels=training_counts.index, autopct='%1.1f%%',
        colors=['lightgreen', 'salmon'], startangle=90)
plt.title('Penerimaan Pelatihan', fontsize=12, fontweight='bold')

# 4.6 Distribusi Wilayah
ax6 = plt.subplot(3, 3, 6)
df['region'].value_counts().plot(kind='bar', color='purple', edgecolor='black')
plt.title('Distribusi Wilayah', fontsize=12, fontweight='bold')
plt.xlabel('Wilayah')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# ============================================
# 5. ANALISIS SKOR KESADARAN & KESIAPAN
# ============================================

print("\n" + "="*50)
print("ANALISIS SKOR KESADARAN & KESIAPAN")
print("="*50)

# Statistik skor berdasarkan kelompok
print("\nRata-rata Skor berdasarkan Penerimaan Pelatihan:")
print(df.groupby('training_received')[['awareness_score', 'preparedness_score']].mean())

print("\nRata-rata Skor berdasarkan Gender:")
print(df.groupby('gender')[['awareness_score', 'preparedness_score']].mean())

print("\nRata-rata Skor berdasarkan Wilayah:")
print(df.groupby('region')[['awareness_score', 'preparedness_score']].mean())

# Visualisasi Skor
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Boxplot Awareness Score by Training
axes[0, 0].boxplot([df[df['training_received']=='Ya']['awareness_score'].values,
                   df[df['training_received']=='Tidak']['awareness_score'].values],
                   labels=['Dengan Pelatihan', 'Tanpa Pelatihan'])
axes[0, 0].set_title('Awareness Score: Dengan vs Tanpa Pelatihan', fontweight='bold')
axes[0, 0].set_ylabel('Skor Kesadaran')
axes[0, 0].grid(True, alpha=0.3)

# Boxplot Preparedness Score by Training
axes[0, 1].boxplot([df[df['training_received']=='Ya']['preparedness_score'].values,
                   df[df['training_received']=='Tidak']['preparedness_score'].values],
                   labels=['Dengan Pelatihan', 'Tanpa Pelatihan'])
axes[0, 1].set_title('Preparedness Score: Dengan vs Tanpa Pelatihan', fontweight='bold')
axes[0, 1].set_ylabel('Skor Kesiapan')
axes[0, 1].grid(True, alpha=0.3)

# Scatter plot Age vs Awareness
axes[1, 0].scatter(df['age'], df['awareness_score'], alpha=0.6, c='blue', edgecolors='black')
axes[1, 0].set_title('Hubungan Umur dengan Awareness Score', fontweight='bold')
axes[1, 0].set_xlabel('Umur')
axes[1, 0].set_ylabel('Awareness Score')
axes[1, 0].grid(True, alpha=0.3)

# Scatter plot Age vs Preparedness
axes[1, 1].scatter(df['age'], df['preparedness_score'], alpha=0.6, c='red', edgecolors='black')
axes[1, 1].set_title('Hubungan Umur dengan Preparedness Score', fontweight='bold')
axes[1, 1].set_xlabel('Umur')
axes[1, 1].set_ylabel('Preparedness Score')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================
# 6. ANALISIS KORELASI
# ============================================

print("\n" + "="*50)
print("ANALISIS KORELASI")
print("="*50)

# Korelasi antara umur, jam pelatihan, dan skor
numeric_cols = ['age', 'training_hours', 'awareness_score', 'preparedness_score']
correlation_matrix = df[numeric_cols].corr()

print("\nMatriks Korelasi:")
print(correlation_matrix)

# Visualisasi korelasi dengan plot sederhana
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(df['training_hours'], df['awareness_score'], alpha=0.6, c='green')
plt.title('Korelasi Jam Pelatihan vs Awareness Score')
plt.xlabel('Jam Pelatihan')
plt.ylabel('Awareness Score')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.scatter(df['training_hours'], df['preparedness_score'], alpha=0.6, c='orange')
plt.title('Korelasi Jam Pelatihan vs Preparedness Score')
plt.xlabel('Jam Pelatihan')
plt.ylabel('Preparedness Score')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================
# 7. ANALISIS BERDASARKAN KELOMPOK UMUR
# ============================================

print("\n" + "="*50)
print("ANALISIS BERDASARKAN KELOMPOK UMUR")
print("="*50)

# Membuat kategori umur
def age_category(age):
    if age <= 12:
        return 'Anak-anak (6-12)'
    elif age <= 18:
        return 'Remaja (13-18)'
    elif age <= 25:
        return 'Dewasa Muda (19-25)'
    elif age <= 35:
        return 'Dewasa (26-35)'
    else:
        return 'Dewasa Tua (36-45)'

df['age_group'] = df['age'].apply(age_category)

# Analisis per kelompok umur
print("\nRata-rata Skor per Kelompok Umur:")
age_group_analysis = df.groupby('age_group')[['awareness_score', 'preparedness_score', 'training_hours']].mean().round(2)
print(age_group_analysis)

print("\nPersentase Penerimaan Pelatihan per Kelompok Umur:")
training_by_age = pd.crosstab(df['age_group'], df['training_received'], normalize='index') * 100
print(training_by_age.round(2))

# Visualisasi per kelompok umur
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Bar plot skor per kelompok umur
age_group_analysis[['awareness_score', 'preparedness_score']].plot(kind='bar', ax=axes[0])
axes[0].set_title('Rata-rata Skor per Kelompok Umur', fontweight='bold')
axes[0].set_xlabel('Kelompok Umur')
axes[0].set_ylabel('Rata-rata Skor')
axes[0].legend(['Awareness', 'Preparedness'])
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True, alpha=0.3)

# Stacked bar untuk training per kelompok umur
training_by_age.plot(kind='bar', stacked=True, ax=axes[1], color=['green', 'red'])
axes[1].set_title('Distribusi Penerimaan Pelatihan per Kelompok Umur', fontweight='bold')
axes[1].set_xlabel('Kelompok Umur')
axes[1].set_ylabel('Persentase')
axes[1].legend(title='Training')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================
# 8. ANALISIS EFEKTIVITAS PELATIHAN
# ============================================

print("\n" + "="*50)
print("ANALISIS EFEKTIVITAS PELATIHAN")
print("="*50)

# Menghitung efektivitas pelatihan
df['effectiveness'] = (df['awareness_score'] + df['preparedness_score']) / 2

print("\nEfektivitas Berdasarkan Status Pelatihan:")
effectiveness_by_training = df.groupby('training_received')['effectiveness'].agg(['mean', 'std', 'min', 'max'])
print(effectiveness_by_training.round(2))

print("\nTop 5 Responden dengan Skor Tertinggi:")
top_5 = df.nlargest(5, 'effectiveness')[['age', 'gender', 'education_level', 'training_hours', 'effectiveness']]
print(top_5)

print("\nBottom 5 Responden dengan Skor Terendah:")
bottom_5 = df.nsmallest(5, 'effectiveness')[['age', 'gender', 'education_level', 'training_hours', 'effectiveness']]
print(bottom_5)

# ============================================
# 9. ANALISIS BERDASARKAN JENIS BENCANA
# ============================================

print("\n" + "="*50)
print("ANALISIS BERDASARKAN JENIS BENCANA")
print("="*50)

# Analisis per jenis bencana
disaster_analysis = df.groupby('disaster_type').agg({
    'awareness_score': 'mean',
    'preparedness_score': 'mean',
    'training_hours': 'mean',
    'age': 'mean'
}).round(2)

print("\nStatistik per Jenis Bencana:")
print(disaster_analysis)

# Visualisasi per jenis bencana
plt.figure(figsize=(12, 6))
disaster_analysis[['awareness_score', 'preparedness_score']].plot(kind='bar')
plt.title('Rata-rata Skor per Jenis Bencana', fontweight='bold', fontsize=14)
plt.xlabel('Jenis Bencana')
plt.ylabel('Rata-rata Skor')
plt.xticks(rotation=45)
plt.legend(['Awareness', 'Preparedness'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================
# 10. SUMMARY & KESIMPULAN
# ============================================

print("\n" + "="*50)
print("RINGKASAN DAN KESIMPULAN")
print("="*50)

print("\nSTATISTIK UMUM:")
print(f"Total Responden: {len(df)}")
print(f"Rentang Umur: {df['age'].min()} - {df['age'].max()} tahun")
print(f"Rata-rata Umur: {df['age'].mean():.1f} tahun")
print(f"Rata-rata Awareness Score: {df['awareness_score'].mean():.1f}")
print(f"Rata-rata Preparedness Score: {df['preparedness_score'].mean():.1f}")
print(f"Rata-rata Jam Pelatihan: {df['training_hours'].mean():.1f} jam")

print("\nPENERIMAAN PELATIHAN:")
training_pct = df['training_received'].value_counts(normalize=True) * 100
print(f"Yang Menerima Pelatihan: {training_pct.get('Ya', 0):.1f}%")
print(f"Yang Tidak Menerima Pelatihan: {training_pct.get('Tidak', 0):.1f}%")

print("\nKESIMPULAN UTAMA:")
# Efektivitas Pelatihan
if df[df['training_received']=='Ya']['awareness_score'].mean() > df[df['training_received']=='Tidak']['awareness_score'].mean():
    print("1. Efektivitas Pelatihan: Pelatihan efektif meningkatkan kesadaran")
else:
    print("1. Efektivitas Pelatihan: Perlu evaluasi metode pelatihan")

# Kelompok Umur dengan Skor Tertinggi
df['effectiveness'] = (df['awareness_score'] + df['preparedness_score']) / 2
best_age_group = df.groupby('age_group')['effectiveness'].mean().idxmax()
print(f"2. Kelompok Umur dengan Skor Tertinggi: {best_age_group}")

# Jenis Bencana dengan Kesadaran Tertinggi
best_disaster = df.groupby('disaster_type')['awareness_score'].mean().idxmax()
print(f"3. Jenis Bencana dengan Kesadaran Tertinggi: {best_disaster}")

# Wilayah dengan Kesiapan Tertinggi
best_region = df.groupby('region')['preparedness_score'].mean().idxmax()
print(f"4. Wilayah dengan Kesiapan Tertinggi: {best_region}")

# Save hasil analisis ke file CSV
print("\nMenyimpan hasil analisis...")

# Simpan beberapa dataframe hasil analisis
age_group_analysis.to_csv('analisis_kelompok_umur.csv')
disaster_analysis.to_csv('analisis_jenis_bencana.csv')
df.to_csv('disaster_processed.csv', index=False)

print("Hasil analisis telah disimpan:")
print(" - analisis_kelompok_umur.csv")
print(" - analisis_jenis_bencana.csv")
print(" - disaster_processed.csv")

print("\n" + "="*50)
print("ANALISIS SELESAI")
print("="*50)
```
