```python
# %%
import pandas as pd
import matplotlib.pyplot as plt

# membaca data
data = pd.read_csv("tips.csv")

# hitung jumlah gender
gender = data['sex'].value_counts()

# membuat donut chart
plt.figure(figsize=(6,6))

plt.pie(
    gender,
    labels=gender.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width':0.4}  # ini yang bikin DONUT
)

plt.title("Sebaran Gender Pemberi Tips (Persentase)")
plt.show()

```

```python
# %%
gender_day = pd.crosstab(data['day'], data['sex'])

gender_day.plot(kind='bar', stacked=True)

plt.title("Perbandingan Gender Pemberi Tip Berdasarkan Hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Pelanggan")

plt.show()
```

```python

# %%
avg_tip = data.groupby('sex')['tip'].mean()

plt.figure()

avg_tip.plot(kind='bar')

plt.title("Rata-rata Tip Berdasarkan Gender")
plt.xlabel("Gender")
plt.ylabel("Rata-rata Tip")

plt.show()
```
